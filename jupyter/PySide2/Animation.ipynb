```python
from PySide2.QtCore import *
from PySide2.QtWidgets import *
from PySide2.QtGui import *
import sys


class Window(QWidget):
    def __init__(self):
        super().__init__()
        self.resize(270, 110)
        layout = QFormLayout()
        layout.setVerticalSpacing(30)
        layout.addRow("Name:", QLineEdit())
        layout.addRow("Job:", QLineEdit())
        emailLabel = QLabel("Email:")
        layout.addRow(emailLabel, QLineEdit())
        self.setLayout(layout)


class Buttons(QWidget):
    def __init__(self):
        super().__init__()
        self.resize(600, 800)
        self.button = QPushButton(text='Push Me ;)', parent=self)
        self.button2 = QPushButton(text='Don\'t Push Me X(', parent=self)
        self.button3 = QPushButton(text='Hi n Bye !', parent=self)

        vbox = QHBoxLayout()
        vbox.addWidget(self.button)
        vbox.addWidget(self.button2)
        vbox.addWidget(self.button3)

        self.setLayout(vbox)
        self.listener()

    def listener(self):
        self.button.clicked.connect(self.for_button)
        self.button2.clicked.connect(self.for_button2)
        self.button3.clicked.connect(self.for_button3)

    @classmethod
    def expand(cls, widget, percent):
        rect1 = widget.geometry()
        left, top, width, height = rect1.left(), rect1.top(), rect1.width(), rect1.height()
        nwidth = width * percent // 100
        nheight = height * percent // 100
        left -= (nwidth - width) // 2
        height -= (nheight - height) // 2
        rect2 = QRect(left, top, nwidth, nheight)
        return rect2

    def for_button(self):
        self.anim = QPropertyAnimation(self.button, b"geometry")
        self.anim.setDuration(1000)
        self.anim.setStartValue(self.button.geometry())
        self.anim.setEndValue(self.expand(self.button, 110))

        curve = QEasingCurve()
        curve.setType(QEasingCurve.OutBounce)
        self.anim.setEasingCurve(curve)
        self.anim.setLoopCount(2)
        self.anim.start()

    def for_button2(self):
        self.anim = QPropertyAnimation(self.button2, b"color")
        self.anim.setDuration(1000)
        self.anim.setStartValue(QColor(0, 0, 0))
        self.anim.setEndValue(QColor(255, 0, 255))

        curve = QEasingCurve()
        curve.setType(QEasingCurve.OutBounce)
        self.anim.setEasingCurve(curve)
        self.anim.setLoopCount(2)
        self.anim.start()

    def for_button3(self):
        self.anim = QPropertyAnimation(self.button3, b"geometry")
        self.anim.setDuration(1000)
        self.anim.setStartValue(self.button3.geometry())
        self.anim.setEndValue(self.expand(self.button3, 105))

        curve = QEasingCurve()
        curve.setType(QEasingCurve.OutBounce)
        self.anim.setEasingCurve(curve)
        self.anim.setLoopCount(2)
        self.anim.start()


class Mainwindow(QMainWindow):
    def __init__(self, parent=None):
        super(Mainwindow, self).__init__(parent)
        self.setGeometry(QRect(100, 100, 800, 600))

        self.button_widget = Buttons()
        self.form_window = Window()

        # self.setCentralWidget(self.form_window)
        self.setCentralWidget(self.button_widget)
        self.show()


if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = Mainwindow()
    sys.exit(app.exec_())
```
