```python
from PySide2.QtCore import *
from PySide2.QtWidgets import *
from PySide2.QtGui import *
import sys


class Paint(QWidget):
    def __init__(self):
        super().__init__()
        self.resize(600, 800)

    def paintEvent(self, event):
        
        # painting text
        self.textPainter = QPainter(self)
        self.textPainter.setFont(QFont('Times', 14, QFont.Bold))
        self.textPainter.drawText(QPoint(20, 30), "Testing !")

        # painting line
        self.linePainter = QPainter(self)
        self.linePainter.drawLine(QPoint(50, 60), QPoint(100, 100))

        # painting rectangle
        self.rectanglePainter = QPainter(self)
        self.rectanglePainter.setBrush(Qt.BDiagPattern)
        self.rectanglePainter.drawRect(QRect(40, 120, 80, 30))

        # painting ellipse
        ellipsePen = QPen()
        ellipsePen.setColor(Qt.red)
        ellipsePen.setStyle(Qt.DashDotLine)

        self.ellipsePainter = QPainter(self)
        self.ellipsePainter.setPen(ellipsePen)
        self.ellipsePainter.drawEllipse(QPoint(80, 200), 50, 20)

        # image attaching
        image = QImage()
        image.load(r"C:\Users\d09am\Music\troll\Jadu.jpg")
        image
        imagePainter = QPainter(self)
        imagePainter.drawImage(QRect(200, 200, 500, 300), image)

class Clock(QWidget):
    def __init__(self):
        super().__init__()
        self.resize(600, 800)
        timer = QTimer(self)
        timer.start(1000)
        timer.timeout.connect(self.update)

    def paintEvent(self, event):
        hour_hand = (
            QPoint(4, 4),
            QPoint(-4, 4),
            QPoint(0, -40),
        )

        minute_hand = (
            QPoint(4, 4),
            QPoint(-4, 4),
            QPoint(0, -70),
        )

        second_hand = (
            QPoint(2, 2),
            QPoint(-2, 2),
            QPoint(0, -90),
        )

        side = min(self.width(), self.height())

        # painter
        painter = QPainter(self)
        painter.setRenderHint(QPainter.Antialiasing)
        painter.translate(self.width() // 2, self.height() // 2)
        painter.scale(side / 250, side / 250)

        # making clock ticks
        for i in range(60):
            if i % 5 != 0:
                painter.drawLine(92, 0, 96, 0)
            else:
                painter.drawLine(86, 0, 96, 0)
            painter.rotate(6.0)

        time = QTime.currentTime()

        # drawing hour hand
        painter.save()
        painter.rotate((time.hour() * 360) // 12)
        painter.setPen(Qt.NoPen)
        painter.setBrush(Qt.black)
        painter.drawConvexPolygon(hour_hand)
        painter.restore()

        # draw minute hand
        painter.save()
        painter.rotate((time.minute() * 360) // 60)
        painter.setPen(Qt.NoPen)
        painter.setBrush(Qt.black)
        painter.drawConvexPolygon(minute_hand)
        painter.restore()

        # draw second had
        painter.save()
        painter.rotate((time.second() * 360) // 60)
        painter.setPen(Qt.NoPen)
        painter.setBrush(Qt.black)
        painter.drawConvexPolygon(second_hand)
        painter.restore()


class Mainwindow(QMainWindow):
    def __init__(self, parent=None):
        super(Mainwindow, self).__init__(parent)
        self.setGeometry(QRect(100, 100, 800, 600))

        self.paint_window = Paint()

        # self.setCentralWidget(self.form_window)
        # self.setCentralWidget(self.button_widget)
        self.setCentralWidget(self.paint_window)
        self.show()


if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = Mainwindow()
    sys.exit(app.exec_())
```
